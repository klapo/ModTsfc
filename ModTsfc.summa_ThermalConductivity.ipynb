```python
# netcdf/numpy/xray/stats
import numpy as np

# import plotting
import seaborn as sns
import matplotlib
from matplotlib.pyplot import subplots
import matplotlib.pyplot as plt
import os

# Customize
sns.set_style("whitegrid")
sns.set_context('paper', font_scale=1.25)
%matplotlib inline
```

```python
lambda_air = 0.026                      # thermal conductivity of air          (W m-1 K-1)
lambda_ice = 2.50                       # thermal conductivity of ice          (W m-1 K-1)
BulkDenIce = np.arange(50, 550)
Yen1965 = 3.217 * 10**(-6) * BulkDenIce**2
Mellor1977 = 2.576 * 10**(-6) * BulkDenIce**2 + 7.4 * 10**(-2)
Jordan1991 = lambda_air + (7.75 * 10**(-5) * BulkDenIce\
                           + 1.105* 10**(-6) * BulkDenIce**2) * (lambda_ice - lambda_air)


fig, ax = plt.subplots(1, 1, figsize=(7, 3))
# Thermal Conductivity(Density)
ax.plot(BulkDenIce, Yen1965, label='Yen')
ax.plot(BulkDenIce, Mellor1977, label='Mellor')
ax.plot(BulkDenIce, Jordan1991, label='Jordan')
# Constant Values
ax.plot([BulkDenIce[0], BulkDenIce[-1]], [0.1, 0.1], '--', color='0.25')
ax.plot([BulkDenIce[0], BulkDenIce[-1]], [0.35, 0.35], '--', color='0.25')
ax.plot([BulkDenIce[0], BulkDenIce[-1]], [1, 1], '--', color='0.25', label='Constant $\lambda$')

# Formatting
ax.set_ylabel('$\lambda$, thermal cond. ($W m^{-1} K^{-1}$)')
ax.set_xlabel('Snow Density ($kg m^{-3}$)')
ax.set_xlim(BulkDenIce[0], BulkDenIce[-1])
ax.set_ylim(0, 1.1)
ax.legend(loc='center left', frameon=True)

os.chdir('/Users/karllapo/gdrive/SnowHydrology/proj/ModTsfc/Graphics')
fig.tight_layout()
fig.savefig('ModTsfc.thermalConductivity.pdf')
```

```python
fig, ax = plt.subplots(1, 1, figsize=(4, 4))
# Thermal Conductivity(Density)
ax.semilogy(BulkDenIce, Yen1965, label='Yen')
ax.semilogy(BulkDenIce, Mellor1977, label='Mellor')
ax.semilogy(BulkDenIce, Jordan1991, label='Jordan')
# Constant Values
ax.semilogy([BulkDenIce[0], BulkDenIce[-1]], [0.1, 0.1], '--', color='0.5')
ax.semilogy([BulkDenIce[0], BulkDenIce[-1]], [0.35, 0.35], '--', color='0.5')
ax.semilogy([BulkDenIce[0], BulkDenIce[-1]], [1.0, 1.0], '--', color='0.5', label='Constant (Smirnova)')

# Formatting
ax.set_ylabel('$\lambda$, thermal cond. ($W m^{-1} K^{-1}$)')
ax.set_xlabel('Snow Density ($kg m^{-3}$)')
ax.set_xlim(BulkDenIce[0], BulkDenIce[-1])
ax.set_ylim(0, 1.1)
ax.legend(loc='lower right', frameon=True)
```
